public class Program
{
    public static double TriangleArea(double a, double b, double c)
    {
        double s = (a + b + c) / 2;
        double area = Math.Sqrt(s * (s - a) * (s - b) * (s - c));
        return area;
    }

    {

A. Write a function inside(x,y,x1,y1,x2,y2) that returns True or False
depending on whether the point (x,y) lies in the rectangle with lower left
corner (x1,y1) and upper right corner (x2,y2).
B. Use function inside() from part a. to write an expression that tests whether
the point (1,1) lies in both of the following rectangles: one with lower left
corner (0.3, 0.5) and upper right corner (1.1, 0.7) and the other with lower
left corner (0.5, 0.2) and upper right corner (1.1, 2).


using System;

public class Program
{
    public static bool Inside(int x, int y, int x1, int y1, int x2, int y2)
    {
        return x >= x1 && x <= x2 && y >= y1 && y <= y2;
    }
    
    public static void Main(string[] args)
    {
        bool rectangle1 = Inside(1, 1, 0, 0, 2, 3);
        bool rectangle2 = Inside(1, 1, 0.3, 0.5, 1.1, 0.7) && Inside(1, 1, 0.5, 0.2, 1.1, 2);
        
        Console.WriteLine(rectangle1);  // Output: True
        Console.WriteLine(rectangle2);  // Output: False
    }
}

```python
## Question 6 Python
## Write a function pig() that takes a word (i.e., a string) as input and returns its pig•Latin form. 
def pig(word):
    vowels = ['a', 'e', 'i', 'o', 'u']
    word = word.lower()
    
    if word[0] in vowels:
        return word + 'ay'
    else:
        return word[1:] + word[0] + 'ay'

# Example usage:
print(pig('happy'))  
print(pig('Enter'))
```

Question 7: Write a function bldcount() that reads the file with name name and reports (i.e., prints) how many patients there are in each bloodtype.

```python
a=open("bloodtype.txt")
data=a.read()
print(data)

bloodlist=data.split()

AB=0
B=0
A=0
O=0
OO=0

for i in range(len(bloodlist)):
    if(bloodlist[i] == 'AB'):
        AB= AB+1
    elif(bloodlist[i] == 'B'):
        B= B+1
    elif(bloodlist[i]=='A'):
        A= A+1
    elif(bloodlist[i]=='O'):
        O= O+1
    elif(bloodlist[i]=='OO'):
        OO=OO+1
        
print("there are {} patients of blood type AB".format(AB))
print("there are {} patients of blood type A".format(A))
print("there are {} patients of blood type B".format(B))
print("there are {} patients of blood type O".format(O))
print("there are {} patients of blood type OO".format(OO))

a.close()
```

```python
def currency_converter(amount, currency):
  conversion_rates = {'AUD': 1.0345157,'CHF':1.0237414,'CNY':0.1550176,'DKK':0.1651442,
               'EUR':1.2296544,'GBP':1.5550989,'HKD':0.1270207, 'INR':0.0177643,
               'JPY': 0.01241401,'MXN': 0.0751848,'MYR': 0.3145411,'NOK':0.1677063,
               'NZD':0.8003591,'PHP':0.0233234,'SEK':0.148269,'SGD':0.788871,'THB':0.0313789}   #Dictionary of currency rates.
  
  conversion_rate = conversion_rates.get(currency, 0)
  if conversion_rate == 0:
    return None
  
  return amount * conversion_rate

amount = float(input("Enter the amount: "))
currency = input("Enter the currency: ")

usd = currency_converter(amount, currency)
if usd is None:
  print("Invalid currency")
else:
  print("{0:.2f}".format(amount), currency, "is equal to", "{0:.2f}".format(usd), "USD")
```

Trying to add incompatible variables, as in adding 6 + 'a':
This will cause a TypeError since Python does not allow addition between an integer and a string.

Referring to the 12th item of a list that has only 10 items:
This will cause an IndexError since the index 12 is out of range for a list with only 10 items. The valid indices for the list would be from 0 to 9.

Using a value that is out of range for a function's input, such as calling math.sqrt(-1.0):
This will cause a ValueError since the math.sqrt() function expects a non-negative value to calculate the square root. Taking the square root of a negative number is mathematically undefined.

Using an undeclared variable, such as print(x) when x has not been defined:
This will cause a NameError since the variable x has not been assigned a value or declared before it is used in the print() statement.

```python
##QUESTION 10

def frequencies(text):
    letter_counts = [0] * 26  # List to store the counts of each letter (a-z)
    
    for char in text.lower():
        if char.isalpha():
            index = ord(char) - ord('a')  # Calculate the index based on ASCII values
            letter_counts[index] += 1
    
    return letter_counts
# Example usage:
result = frequencies("Poonam")
print(result)
```
